```python
dummies=pd.get_dummies(df.city)
dum_animal=pd.get_dummies(df.animal)
dum_furnit =pd.get_dummies(df.furniture)
```

```python
merged=pd.concat([df,dum_furnit,dum_animal, pd.get_dummies(df.city)],axis='columns')
merged
```

```python
final=merged.drop(['city','Porto Alegre','animal','furniture','not furnished','not acept'],axis='columns')
final
```

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
reg=linear_model.LinearRegression()
reg.fit(final[['rooms','bathroom','parking spaces','floor','hoa (R$)','property tax (R$)','fire insurance (R$)']],final[['rent amount (R$)']] )
```
